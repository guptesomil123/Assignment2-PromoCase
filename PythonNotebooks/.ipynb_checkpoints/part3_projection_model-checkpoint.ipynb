# Part 3 — Simple Projection Model

**Goal:** Project this year's SPRING20 promo (3-day event) using last year's  
performance as an anchor, with a justified low / mid / high range.

---

## Data Honesty Statement

Before any numbers, here is exactly what the data can and cannot support:

| What we want to project | Data available | Method |
|---|---|---|
| TY projected **units** | ✅ LY SKU-level units in `last_year_promo_skus.csv` | LY promo units × YoY growth |
| TY projected **orders** | ⚠️ No LY order count | Estimated: projected units ÷ TY baseline UPT |
| TY projected **revenue** | ⚠️ No LY revenue | Estimated: projected orders × TY baseline AOV |

This is a **units-anchored model.** Revenue and orders are derived from TY  
assumptions (AOV, UPT from Part 2 baseline) — not directly from LY data.  
That introduces two additional assumption layers, each documented in Section 8.

> *"If a number feels arbitrary, say so. We'd rather see you flag uncertainty  
> than launder it through false precision."* — Assignment brief


## 1. Imports & Scenario Parameters

````python
import pandas as pd
import numpy as np

pd.set_option("display.float_format", "{:,.2f}".format)

# ── Scenario parameters ────────────────────────────────────────────────────────
# Full justification in Section 4. Summary:
#   Low  1.10x — DTC decelerating as retail channel (Vitamin Shoppe) absorbs new demand
#   Mid  1.18x — Assignment input; consistent with DTC share of 1.50x total 2025 growth
#   High 1.25x — Upper bound of DTC portion; $250M projection proved overstated (actual: $150M)

GROWTH = {
    "low" : 1.10,
    "mid" : 1.18,
    "high": 1.25,
}

# Uplift scalar: did THIS promo over/underperform vs LY promo (relative to its own baseline)?
UPLIFT_SCALAR = {
    "low" : 0.85,   # Promo fatigue, more competitive promos in market
    "mid" : 1.00,   # Repeat performance — same relative uplift as LY
    "high": 1.10,   # Stronger awareness, larger email/SMS list YoY
}

# Promo AOV adjustment: 20% discount compresses AOV, partially offset by larger baskets
PROMO_AOV_ADJ = {
    "low" : 0.82,   # Deep compression — customers cherry-pick lower-price items
    "mid" : 0.87,   # Moderate — standard DTC assumption for sitewide % discount
    "high": 0.92,   # Mild — customers buy as usual, just at a discount
}

PROMO_DAYS = 3

print("=== SCENARIO PARAMETERS ===")
print(f"{'Scenario':<10} {'Growth':>8} {'Uplift':>8} {'AOV adj':>9} {'Combined':>10}")
print("-" * 50)
for s in ["low", "mid", "high"]:
    print(f"{s:<10} {GROWTH[s]:>7.2f}x {UPLIFT_SCALAR[s]:>7.2f}x "
          f"{PROMO_AOV_ADJ[s]:>8.2f}x {GROWTH[s]*UPLIFT_SCALAR[s]:>9.3f}x")
````

## 2. Load Data

Three inputs required:
- `last_year_promo_skus.csv` — LY SKU-level units by phase (our anchor)
- `orders_clean.csv` — TY orders for baseline AOV and UPT


````python
ly = pd.read_csv("/Users/apple/Desktop/Assignment2/Source Data/last_year_promo_skus.csv")
ly.columns = ly.columns.str.strip().str.lower().str.replace(" ", "_")

print("=== LAST YEAR PROMO SKU DATA ===")
print(f"Shape  : {ly.shape}")
print(f"Phases : {sorted(ly['phase'].unique())}")
print(f"SKUs   : {ly['sku'].nunique()}")
print()
ly.head(8)
````

````python
df = pd.read_csv("/Users/apple/Desktop/Assignment2/Output/orders_clean.csv", parse_dates=["order_date"])

BASELINE_START = pd.Timestamp("2025-02-28")
BASELINE_END   = pd.Timestamp("2025-04-24")
PROMO_START    = pd.Timestamp("2025-04-25")
PROMO_END      = pd.Timestamp("2025-04-27")
PROMO_CODE     = "SPRING20"
BASELINE_DAYS  = (BASELINE_END - BASELINE_START).days + 1

baseline_df = df[
    (df["order_date"] >= BASELINE_START) &
    (df["order_date"] <= BASELINE_END)
]

TY_BASELINE_AOV  = baseline_df["net_sales"].sum() / len(baseline_df)
TY_BASELINE_UPT  = baseline_df["units"].sum()     / len(baseline_df)
TY_DAILY_AVG_ORD = len(baseline_df) / BASELINE_DAYS
TY_DAILY_AVG_REV = baseline_df["net_sales"].sum() / BASELINE_DAYS

print("=== TY BASELINE STATS (Feb 28 – Apr 24 2025) ===")
print(f"  Days in window        : {BASELINE_DAYS}")
print(f"  Avg daily orders      : {TY_DAILY_AVG_ORD:.1f}")
print(f"  Avg daily revenue ($) : {TY_DAILY_AVG_REV:,.2f}")
print(f"  Baseline AOV ($)      : {TY_BASELINE_AOV:.2f}")
print(f"  Baseline UPT          : {TY_BASELINE_UPT:.2f}")
````

## 3. Extract Last Year's Promo Anchor

We extract **promo week units** per SKU from `last_year_promo_skus.csv`.  
We also compute the **LY uplift factor** per SKU:

```
LY uplift factor = LY promo week units ÷ LY pre-promo avg weekly units
```

This tells us how much each SKU was amplified by the promo last year,  
which feeds into the SKU-level work in Parts 5–6.  
For Part 3 we use the aggregate total.


````python
ly_prepromo = ly[ly["phase"] == "Pre-promo"]
ly_promo    = ly[ly["phase"] == "Promo week"]
ly_post     = ly[ly["phase"] == "Post-promo"]

# LY pre-promo avg per SKU (mean of W15 and W16)
ly_pre_avg = (
    ly_prepromo
    .groupby("sku")["units_sold"]
    .mean()
    .reset_index()
    .rename(columns={"units_sold": "ly_pre_avg_units"})
)

ly_promo_units = (
    ly_promo[["sku", "units_sold"]]
    .rename(columns={"units_sold": "ly_promo_units"})
)

ly_sku = ly_pre_avg.merge(ly_promo_units, on="sku")
ly_sku["ly_uplift_factor"] = ly_sku["ly_promo_units"] / ly_sku["ly_pre_avg_units"]

print("=== LY SKU-LEVEL UPLIFT FACTORS ===")
print(ly_sku.sort_values("ly_uplift_factor", ascending=False).to_string(index=False))
````

````python
LY_TOTAL_PROMO_UNITS = ly_sku["ly_promo_units"].sum()
LY_AVG_UPLIFT        = ly_sku["ly_uplift_factor"].mean()

print(f"LY total promo week units  : {LY_TOTAL_PROMO_UNITS:,.0f}")
print(f"LY avg uplift factor       : {LY_AVG_UPLIFT:.2f}x")
print(f"LY uplift range            : {ly_sku['ly_uplift_factor'].min():.2f}x – {ly_sku['ly_uplift_factor'].max():.2f}x")
print()
print("Note: uplift varies meaningfully by SKU.")
print("Parts 5–6 handle SKU-level forecasting. Here we use the aggregate.")
````

## 4. Justifying the Growth Multiplier

The assignment provides **1.18x** as the YoY baseline growth rate and explicitly  
asks us to interrogate it. Here is our reasoning, grounded in external data.

### Cymbiotika's actual revenue trajectory

| Year | Revenue | YoY growth |
|---|---|---|
| 2022 | $41M | — |
| 2024 | $100M+ | ~1.56x/yr implied |
| 2025 | **$150M** (actual) | **1.50x** |

Source: LinkedIn / Google AI overview (Kunle Campbell, multiple sources, 2025).

### Why 1.18x is still the right mid-case for DTC

Total company revenue grew 1.50x in 2025 — but not all of that was DTC:

- Cymbiotika entered **The Vitamin Shoppe in 2024**, adding a retail channel  
  that absorbs a meaningful share of new customer acquisition.
- Subscribers (who drive DTC repeat revenue) have **6+ month retention** —  
  the DTC base is sticky but slower-growing than headline revenue implies.
- A LinkedIn article projected $250M for 2025; **actual came in at $150M**,  
  suggesting aggressive growth assumptions should be treated with caution.

If roughly half of the 1.50x total growth was driven by retail expansion,  
the DTC-specific growth rate is closer to **1.20–1.25x**. The assignment's  
1.18x is therefore a conservative but defensible mid-case for DTC.

### Scenario rationale

| Scenario | Multiplier | Reasoning |
|---|---|---|
| **Low** | 1.10x | DTC decelerating as retail absorbs new demand; promo fatigue |
| **Mid** | 1.18x | Assignment input; consistent with DTC share of 1.50x total growth |
| **High** | 1.25x | Upper DTC bound; $250M projection proved overstated so we cap here |


````python
# Sensitivity table — what different growth rates imply for projected units
print("=== SENSITIVITY: GROWTH RATE vs PROJECTED TOTAL PROMO UNITS ===")
print(f"{'Growth':>8}  {'Proj units':>12}  {'vs LY units':>12}  Note")
print("-" * 60)
for g in [1.05, 1.10, 1.15, 1.18, 1.20, 1.25, 1.30, 1.50]:
    proj  = LY_TOTAL_PROMO_UNITS * g
    delta = proj - LY_TOTAL_PROMO_UNITS
    notes = {1.10: "← low case", 1.18: "← mid (assignment input)",
             1.25: "← high case", 1.50: "← total co. growth (incl. retail)"}
    note = notes.get(g, "")
    print(f"  {g:.2f}x   {proj:>12,.0f}   +{delta:>10,.0f}  {note}")
````

## 5. Build Low / Mid / High Projections

### Two-step model

**Step 1 — Project units (LY anchor):**
```
TY projected units = LY promo units × YoY growth × uplift scalar
```

**Step 2 — Convert to orders & revenue (TY assumptions):**
```
TY projected orders  = projected units ÷ TY baseline UPT
TY projected revenue = projected orders × (TY baseline AOV × promo AOV adjustment)
```

### Why use TY baseline for AOV/UPT?
No LY order-level data exists in the provided files — only LY units by SKU.  
TY baseline is the best available proxy. Limitations are documented in Section 8.

### Why adjust AOV downward?
A 20% sitewide discount does not reduce AOV by 20%. Customers often buy  
more units to "make the most of" the discount, partially offsetting the  
discount effect. Net AOV compression is typically 8–18% for a 20% off event.


````python
results = {}

for s in ["low", "mid", "high"]:
    g       = GROWTH[s]
    u       = UPLIFT_SCALAR[s]
    aov_adj = PROMO_AOV_ADJ[s]

    proj_units   = LY_TOTAL_PROMO_UNITS * g * u
    proj_orders  = proj_units / TY_BASELINE_UPT
    proj_aov     = TY_BASELINE_AOV * aov_adj
    proj_revenue = proj_orders * proj_aov

    results[s] = {
        "YoY growth multiplier"      : g,
        "Uplift scalar"              : u,
        "AOV adjustment"             : aov_adj,
        "Combined unit multiplier"   : round(g * u, 3),
        "Projected total units"      : round(proj_units),
        "Projected total orders"     : round(proj_orders),
        "Projected promo AOV ($)"    : round(proj_aov, 2),
        "Projected total revenue ($)": round(proj_revenue, 2),
        "Proj avg daily orders"      : round(proj_orders / PROMO_DAYS, 1),
        "Proj avg daily revenue ($)" : round(proj_revenue / PROMO_DAYS, 2),
    }

print("Projections built for all three scenarios.")
````

## 6. Projection Summary Table

````python
# TY actuals — for post-event validation
ty_promo_narrow = df[
    (df["order_date"] >= PROMO_START) &
    (df["order_date"] <= PROMO_END) &
    (df["discount_code"] == PROMO_CODE)
]

actuals = {
    "Projected total units"      : ty_promo_narrow["units"].sum(),
    "Projected total orders"     : len(ty_promo_narrow),
    "Projected promo AOV ($)"    : ty_promo_narrow["net_sales"].mean(),
    "Projected total revenue ($)": ty_promo_narrow["net_sales"].sum(),
}

metrics = [
    ("Total units",    "Projected total units",       "{:>10,.0f}"),
    ("Total orders",   "Projected total orders",      "{:>10,.0f}"),
    ("Promo AOV ($)",  "Projected promo AOV ($)",     "{:>10,.2f}"),
    ("Total rev ($)",  "Projected total revenue ($)", "{:>10,.0f}"),
]

print("=" * 72)
print(f"{'Metric':<20} {'Low':>10} {'Mid':>10} {'High':>10} {'Actual TY':>10}")
print(f"{'':20} {'(1.10x)':>10} {'(1.18x)':>10} {'(1.25x)':>10} {'(actual)':>10}")
print("=" * 72)
for label, key, fmt in metrics:
    lo  = fmt.format(results["low"][key])
    mid = fmt.format(results["mid"][key])
    hi  = fmt.format(results["high"][key])
    act = fmt.format(actuals[key])
    print(f"{label:<20} {lo} {mid} {hi} {act}")
print("=" * 72)
print()
print("* Actual TY shown for post-event model validation only.")
print("  A true pre-event forecast would not have this column.")
````

````python
# Daily averages summary
print("=== DAILY AVERAGES DURING PROMO (3-day event) ===")
print(f"{'Metric':<28} {'Low':>10} {'Mid':>10} {'High':>10} {'Actual':>10}")
print("-" * 65)

daily_metrics = [
    ("Avg daily orders",    "Proj avg daily orders",     "{:>10.1f}"),
    ("Avg daily rev ($)",   "Proj avg daily revenue ($)","${:>9,.0f}"),
]

actual_daily = {
    "Proj avg daily orders"     : len(ty_promo_narrow) / PROMO_DAYS,
    "Proj avg daily revenue ($)": ty_promo_narrow["net_sales"].sum() / PROMO_DAYS,
}

for label, key, fmt in daily_metrics:
    lo  = fmt.format(results["low"][key])
    mid = fmt.format(results["mid"][key])
    hi  = fmt.format(results["high"][key])
    act = fmt.format(actual_daily[key])
    print(f"{label:<28} {lo} {mid} {hi} {act}")
````

## 7. Post-Event Accuracy Check

Since the promo has already happened, we can validate the model against actuals.  
This is what a real analyst would do after the event to calibrate future forecasts.  
The scenario that lands closest informs which assumptions were most accurate.


````python
print("=== MODEL ACCURACY vs TY ACTUALS ===")
print()

for s in ["low", "mid", "high"]:
    print(f"--- {s.upper()} SCENARIO (growth={GROWTH[s]:.2f}x, uplift={UPLIFT_SCALAR[s]:.2f}x) ---")
    for label, key, fmt in metrics:
        proj   = results[s][key]
        actual = actuals[key]
        err    = proj - actual
        pct    = (err / actual * 100) if actual != 0 else np.nan
        flag   = " ✓" if abs(pct) <= 10 else (" ⚠" if abs(pct) <= 20 else " ✗")
        print(f"  {label:<20}: proj={fmt.format(proj)}  actual={fmt.format(actual)}  "
              f"error={pct:>+6.1f}%{flag}")
    print()

print("Legend:  ✓ within ±10%   ⚠ within ±20%   ✗ off by >20%")
````

## 8. Assumption Log

Every assumption named, justified, and stress-tested.

---

### Assumption 1: LY promo week units are the right anchor
**Assumed:** W17 2024 units = last year's promo demand.  
**Justified:** It is the only direct LY promo data available.  
**Breaks if:** LY W17 had an unusual factor (stockout, viral moment, competitor  
outage) that won't repeat. No way to check from available data.

---

### Assumption 2: YoY growth of 1.10x / 1.18x / 1.25x
**Assumed:** DTC-specific baseline growth in this range.  
**Justified:** Cymbiotika grew total revenue from $100M (2024) to $150M (2025) =  
1.50x overall. Retail expansion (Vitamin Shoppe, 2024) absorbed a meaningful share  
of that growth, putting DTC-specific growth at ~1.18–1.25x. The LinkedIn article  
projected $250M for 2025 — actual came in at $150M — so the 1.25x high case is  
already capped conservatively relative to that overstatement.  
**Breaks if:** The retail/DTC split is materially different from assumed, or if a  
major product launch or channel event skewed 2024 vs 2025 DTC comparability.

---

### Assumption 3: TY promo UPT ≈ TY baseline UPT
**Assumed:** Customers buy the same number of units per order during the promo  
as on a normal baseline day.  
**Breaks if:** Promo events typically drive higher UPT — customers buy more to  
"make the most of" the discount. If true, projected order count is overstated  
(same units distributed across fewer, larger orders).

---

### Assumption 4: Promo AOV = baseline AOV × 0.82–0.92
**Assumed:** 20% discount compresses AOV by 8–18% net.  
**Justified:** Standard DTC assumption for sitewide percentage discounts.  
**Breaks if:** Customers cherry-pick only low-price items (deeper compression)  
or if the promo drives large bundle purchases (milder compression).

---

### Assumption 5: Uplift scalar is uniform across all SKUs
**Assumed:** Growth and uplift apply equally in aggregate.  
**Breaks if:** Hero SKUs grow at a different rate than slow-movers — the aggregate  
masks divergence. SKU-level treatment is in Parts 5–6.


## 9. Save Outputs

````python
out_rows = []
for s in ["low", "mid", "high"]:
    row = {"scenario": s}
    row.update(results[s])
    out_rows.append(row)

out_df = pd.DataFrame(out_rows).set_index("scenario")
out_df.to_csv("/Users/apple/Desktop/Assignment2/Output/part3_projection.csv")
print("Saved → part3_projection.csv")
print()
print("=== FINAL OUTPUT ===")
print(out_df[[
    "Projected total units",
    "Projected total orders",
    "Projected promo AOV ($)",
    "Projected total revenue ($)",
]].to_string())
````

## 10. Key Takeaways

1. **Mid-case (1.18x)** uses the given growth rate and assumes repeat LY uplift  
   performance. It is the most conservative defensible estimate.

2. **Revenue is not directly anchored to LY** — we have no LY revenue data.  
   It is derived from projected orders × adjusted TY AOV. This is the weakest  
   link in the model and should be stated explicitly to Finance.

3. **The range is real, not decorative.** Low–high spread reflects genuine  
   uncertainty in growth rate, uplift repeatability, and promo-day AOV behaviour.

4. **External data supports the multiplier range.** Cymbiotika's actual 2025  
   revenue ($150M vs $100M) provides an external anchor for the growth rate.  
   The $250M LinkedIn projection proved overstated, which is why we capped  
   the high case at 1.25x rather than 1.30x.

5. **Section 7 shows model accuracy.** The closest scenario tells you which  
   assumptions held — use that to calibrate the Part 5 forecasting model.
