# Part 5 — Forecasting Model Build

**Purpose:** Build the forecast Finance will use for next quarter's SPRING20 promo planning.

**What we're testing:** Whether the model is something Finance can actually plan against —
clear logic, explicit assumptions, sensitivity analysis, and honest communication of risk.

---

## Model Architecture

This is a **two-anchor blended forecast**. Rather than relying on a single multiplier,
we cross-check two independent signals and blend them:

- **Anchor A — LY event:** Last year's SPRING20 performance scaled by YoY growth.
  Captures event-specific demand patterns (customer excitement, promo awareness).
- **Anchor B — TY baseline:** This year's pre-promo daily run rate scaled by LY uplift factor.
  Captures current business momentum independent of last year's event.

```
Forecast = (weight_A × Anchor_A) + (weight_B × Anchor_B)
```

Blending guards against over-relying on either anchor alone. If LY had an anomaly
(stockout, viral moment), Anchor A would mislead. If TY baseline drifted from trend,
Anchor B would mislead. The blend dampens both risks.

---

## Inputs Summary

| Input | Value | Source |
|---|---|---|
| LY SPRING20 orders | 268 | TY actuals × 0.75 (assignment assumption) |
| LY SPRING20 net sales | $26,721 | TY actuals × 0.75 |
| LY AOV | $99.70 | LY net sales ÷ LY orders |
| TY baseline avg daily orders | 70.8 | Part 2 — 8-week trailing avg |
| TY baseline avg daily revenue | $8,371 | Part 2 — 8-week trailing avg |
| TY baseline AOV | $118.31 | Part 2 |
| TY baseline UPT | 2.05 | Part 2 |
| Promo duration | 3 days | Given |
| Promo type | 20% off sitewide | Given |


## 1. Imports & Confirmed Inputs

````python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

pd.set_option("display.float_format", "{:,.2f}".format)

# ── Confirmed inputs from data ─────────────────────────────────────────────────
# These values were derived by running the diagnostic snippet on orders_clean.csv

PROMO_DAYS = 3

# TY actuals (SPRING20 Apr 25-27 2025)
TY_PROMO_ORDERS   = 357
TY_PROMO_REVENUE  = 35_627.48
TY_PROMO_AOV      = 99.80
TY_PROMO_UPT      = 2.11

# TY baseline (8-week trailing, Feb 28 – Apr 24 2025)
TY_BASE_DAILY_ORDERS  = 70.8
TY_BASE_DAILY_REVENUE = 8_370.55
TY_BASE_AOV           = 118.31
TY_BASE_UPT           = 2.05

# LY derived (TY × 0.75 — assignment assumption)
LY_PROMO_ORDERS   = round(TY_PROMO_ORDERS  * 0.75)   # 268
LY_PROMO_REVENUE  = TY_PROMO_REVENUE * 0.75           # $26,720.61
LY_PROMO_AOV      = LY_PROMO_REVENUE / LY_PROMO_ORDERS

print("=== CONFIRMED INPUTS ===")
print(f"TY promo orders        : {TY_PROMO_ORDERS}")
print(f"TY promo revenue       : ${TY_PROMO_REVENUE:,.2f}")
print(f"TY promo AOV           : ${TY_PROMO_AOV:.2f}")
print(f"TY promo UPT           : {TY_PROMO_UPT:.2f}")
print()
print(f"TY baseline daily ord  : {TY_BASE_DAILY_ORDERS}")
print(f"TY baseline daily rev  : ${TY_BASE_DAILY_REVENUE:,.2f}")
print(f"TY baseline AOV        : ${TY_BASE_AOV:.2f}")
print(f"TY baseline UPT        : {TY_BASE_UPT:.2f}")
print()
print(f"LY promo orders (×0.75): {LY_PROMO_ORDERS}")
print(f"LY promo revenue       : ${LY_PROMO_REVENUE:,.2f}")
print(f"LY promo AOV           : ${LY_PROMO_AOV:.2f}")
````

## 2. Derive the LY Uplift Factor

The LY uplift factor tells us how much the promo amplified demand relative to
last year's baseline. We need LY baseline daily orders to compute this.

Since we don't have LY order-level data, we back-calculate LY baseline from
TY baseline using the same 0.75 scaling assumption.

**DECISION:** Apply the 0.75 scaling to baseline as well as promo.
Rationale: if LY promo was 75% of TY promo volume, it's reasonable to assume
LY baseline was also approximately 75% of TY baseline — the business was simply
smaller across the board last year.


````python
# LY baseline (derived)
LY_BASE_DAILY_ORDERS  = TY_BASE_DAILY_ORDERS  * 0.75
LY_BASE_DAILY_REVENUE = TY_BASE_DAILY_REVENUE * 0.75

# LY promo daily averages
LY_PROMO_DAILY_ORDERS  = LY_PROMO_ORDERS  / PROMO_DAYS
LY_PROMO_DAILY_REVENUE = LY_PROMO_REVENUE / PROMO_DAYS

# LY uplift factors
LY_ORDER_UPLIFT  = LY_PROMO_DAILY_ORDERS  / LY_BASE_DAILY_ORDERS
LY_REVENUE_UPLIFT = LY_PROMO_DAILY_REVENUE / LY_BASE_DAILY_REVENUE

print("=== LY DERIVED BASELINE ===")
print(f"LY baseline daily orders  : {LY_BASE_DAILY_ORDERS:.1f}")
print(f"LY baseline daily revenue : ${LY_BASE_DAILY_REVENUE:,.2f}")
print()
print("=== LY UPLIFT FACTORS ===")
print(f"LY order uplift factor    : {LY_ORDER_UPLIFT:.2f}x")
print(f"LY revenue uplift factor  : {LY_REVENUE_UPLIFT:.2f}x")
print()
print("Interpretation: the SPRING20 promo last year drove")
print(f"{LY_ORDER_UPLIFT:.1f}x the daily order rate and")
print(f"{LY_REVENUE_UPLIFT:.1f}x the daily revenue of a normal baseline day.")
````

## 3. Promo AOV Adjustment

The 20% sitewide discount compresses AOV. But customers often buy more units
to "make the most of" the discount, partially offsetting the compression.

**Observed this year:** TY promo AOV = $99.80 vs baseline AOV = $118.31.
That is a **15.7% compression** — right in the expected 8–18% range.
TY promo UPT = 2.11 vs baseline UPT = 2.05 — confirming the larger basket offset.

We use this observed compression as the mid-case and build a range around it.


````python
# Observed TY AOV compression
observed_aov_compression = (TY_PROMO_AOV - TY_BASE_AOV) / TY_BASE_AOV
print(f"Observed TY AOV compression : {observed_aov_compression:.1%}")
print(f"  Baseline AOV : ${TY_BASE_AOV:.2f}")
print(f"  Promo AOV    : ${TY_PROMO_AOV:.2f}")
print()

# AOV adjustment factors per scenario
# Low: deeper compression (customers trade down, cherry-pick low-price items)
# Mid: same compression as observed this year
# High: milder compression (customers buy higher-value bundles)
AOV_ADJ = {
    "low" : 0.80,
    "mid" : round(1 + observed_aov_compression, 4),   # observed: ~0.844
    "high": 0.92,
}

print("=== AOV ADJUSTMENT FACTORS PER SCENARIO ===")
for s, adj in AOV_ADJ.items():
    print(f"  {s:<6}: {adj:.3f}x → projected promo AOV = ${TY_BASE_AOV * adj:.2f}")
````

## 4. Scenario Parameters

Three scenarios differ on two dimensions:

**Growth rate (YoY):** How much has DTC demand grown since last year?
Grounded in external data: Cymbiotika grew total revenue from $100M (2024) to
$150M (2025) = 1.50x overall. Retail expansion absorbed a share of that growth,
putting DTC-specific growth at ~1.18–1.25x. The assignment provides 1.18 as
the given rate; we use it as mid and build a defensible range around it.

**Uplift repeatability:** Will this promo drive the same relative lift as last year?
Low assumes some fatigue or competitive pressure; high assumes improved awareness.


````python
# YoY growth multipliers (applied to LY anchor)
GROWTH = {
    "low" : 1.10,   # DTC decelerating as retail channel absorbs new demand
    "mid" : 1.18,   # Assignment input; consistent with DTC share of 1.50x total growth
    "high": 1.25,   # Upper DTC bound; $250M projection proved overstated (actual: $150M)
}

# Uplift repeatability scalars
UPLIFT_SCALAR = {
    "low" : 0.85,   # Promo fatigue, more competitive promos in market
    "mid" : 1.00,   # Repeat performance — same relative lift as LY
    "high": 1.10,   # Stronger awareness, larger email/SMS list YoY
}

# Blend weights: Anchor A (LY event) vs Anchor B (TY baseline × LY uplift)
BLEND = {"anchor_A": 0.60, "anchor_B": 0.40}

print("=== SCENARIO PARAMETERS ===")
print(f"{'Scenario':<10} {'Growth':>8} {'Uplift':>8} {'AOV adj':>9} {'Combined':>10}")
print("-" * 50)
for s in ["low","mid","high"]:
    print(f"{s:<10} {GROWTH[s]:>7.2f}x {UPLIFT_SCALAR[s]:>7.2f}x "
          f"{AOV_ADJ[s]:>8.3f}x {GROWTH[s]*UPLIFT_SCALAR[s]:>9.3f}x")
print()
print(f"Blend weights: {BLEND['anchor_A']:.0%} LY event anchor + "
      f"{BLEND['anchor_B']:.0%} TY baseline anchor")
````

## 5. Build the Forecast — Two-Anchor Blended Model

### Formula

**Anchor A — LY event scaled:**
```
A_orders  = LY promo daily orders × YoY growth × uplift scalar
A_revenue = LY promo daily revenue × YoY growth × uplift scalar × (promo AOV adj / LY AOV compression)
```

**Anchor B — TY baseline scaled:**
```
B_orders  = TY baseline daily orders × LY order uplift × uplift scalar
B_revenue = TY baseline daily revenue × LY revenue uplift × uplift scalar × AOV adj factor
```

**Blended daily forecast:**
```
daily_orders  = 0.60 × A_orders  + 0.40 × B_orders
daily_revenue = 0.60 × A_revenue + 0.40 × B_revenue
```

**Total 3-day forecast:**
```
total_orders  = daily_orders  × 3
total_revenue = daily_revenue × 3
```


````python
results = {}

for s in ["low", "mid", "high"]:
    g   = GROWTH[s]
    u   = UPLIFT_SCALAR[s]
    aov = AOV_ADJ[s]

    # Anchor A: LY event scaled by growth and uplift
    A_orders  = LY_PROMO_DAILY_ORDERS  * g * u
    A_revenue = LY_PROMO_DAILY_REVENUE * g * u * (aov / (1 + observed_aov_compression))

    # Anchor B: TY baseline × LY uplift factor × uplift scalar
    B_orders  = TY_BASE_DAILY_ORDERS  * LY_ORDER_UPLIFT  * u
    B_revenue = TY_BASE_DAILY_REVENUE * LY_REVENUE_UPLIFT * u * aov

    # Blend
    daily_orders  = BLEND["anchor_A"] * A_orders  + BLEND["anchor_B"] * B_orders
    daily_revenue = BLEND["anchor_A"] * A_revenue + BLEND["anchor_B"] * B_revenue

    total_orders  = daily_orders  * PROMO_DAYS
    total_revenue = daily_revenue * PROMO_DAYS
    proj_aov      = total_revenue / total_orders
    proj_upt      = total_orders * (TY_BASE_UPT * 1.03) / total_orders  # slight basket increase

    results[s] = {
        "growth_multiplier"      : g,
        "uplift_scalar"          : u,
        "aov_adjustment"         : aov,
        "anchor_A_daily_orders"  : round(A_orders, 1),
        "anchor_B_daily_orders"  : round(B_orders, 1),
        "blended_daily_orders"   : round(daily_orders, 1),
        "blended_daily_revenue"  : round(daily_revenue, 2),
        "total_orders"           : round(total_orders),
        "total_revenue"          : round(total_revenue, 2),
        "projected_AOV"          : round(proj_aov, 2),
    }

print("Forecasts computed for all three scenarios.")
````

## 6. Forecast Summary Table

````python
print("=" * 76)
print(f"{'Metric':<30} {'Low (1.10x)':>12} {'Mid (1.18x)':>12} {'High (1.25x)':>12} {'TY Actual':>12}")
print("=" * 76)

rows = [
    ("Total orders",         "total_orders",         "{:>12,.0f}"),
    ("Total net sales ($)",  "total_revenue",        "{:>12,.0f}"),
    ("Projected AOV ($)",    "projected_AOV",        "{:>12,.2f}"),
    ("Avg daily orders",     "blended_daily_orders", "{:>12,.1f}"),
    ("Avg daily revenue ($)","blended_daily_revenue","{:>12,.0f}"),
]

actuals = {
    "total_orders"          : TY_PROMO_ORDERS,
    "total_revenue"         : TY_PROMO_REVENUE,
    "projected_AOV"         : TY_PROMO_AOV,
    "blended_daily_orders"  : TY_PROMO_ORDERS / PROMO_DAYS,
    "blended_daily_revenue" : TY_PROMO_REVENUE / PROMO_DAYS,
}

for label, key, fmt in rows:
    lo  = fmt.format(results["low"][key])
    mid = fmt.format(results["mid"][key])
    hi  = fmt.format(results["high"][key])
    act = fmt.format(actuals[key])
    print(f"{label:<30} {lo} {mid} {hi} {act}")

print("=" * 76)
print()
print("* TY Actual shown for post-event model validation only.")
print("  A real pre-event forecast would not have this column.")
````

## 7. Post-Event Model Accuracy

Since we have TY actuals, we can validate how well each scenario performed.
This calibrates the model for future promos.


````python
print("=== MODEL ACCURACY vs TY ACTUALS ===")
print()
accuracy_metrics = [
    ("Total orders",    "total_orders",   TY_PROMO_ORDERS),
    ("Total revenue",   "total_revenue",  TY_PROMO_REVENUE),
    ("AOV",             "projected_AOV",  TY_PROMO_AOV),
]
for s in ["low", "mid", "high"]:
    print(f"--- {s.upper()} SCENARIO ---")
    for label, key, actual in accuracy_metrics:
        proj = results[s][key]
        err  = (proj - actual) / actual * 100
        flag = "✓" if abs(err) <= 10 else ("⚠" if abs(err) <= 20 else "✗")
        print(f"  {label:<16}: proj={proj:>10,.1f}  actual={actual:>10,.1f}  "
              f"error={err:>+6.1f}%  {flag}")
    print()
print("Legend: ✓ within ±10%   ⚠ within ±20%   ✗ off by >20%")
````

## 8. Sensitivity Analysis

**Which assumption moves the forecast most?**

We test each assumption independently by varying it ±10% while holding all others
at mid-case values. The assumption with the largest impact on total revenue is the
one Finance should spend the most time stress-testing.


````python
BASE = results["mid"]["total_revenue"]
assumptions = {
    "YoY growth rate"    : ("growth",  GROWTH["mid"],          0.10),
    "Uplift scalar"      : ("uplift",  UPLIFT_SCALAR["mid"],   0.10),
    "AOV adjustment"     : ("aov",     AOV_ADJ["mid"],         0.10),
    "LY order uplift"    : ("ly_up",   LY_ORDER_UPLIFT,        0.10),
    "Blend weight (A/B)" : ("blend",   BLEND["anchor_A"],      0.10),
}

def run_single(vary_param, vary_val):
    g   = vary_val if vary_param == "growth" else GROWTH["mid"]
    u   = vary_val if vary_param == "uplift" else UPLIFT_SCALAR["mid"]
    aov = vary_val if vary_param == "aov"    else AOV_ADJ["mid"]
    ly  = vary_val if vary_param == "ly_up"  else LY_ORDER_UPLIFT
    bw  = vary_val if vary_param == "blend"  else BLEND["anchor_A"]

    A_orders  = LY_PROMO_DAILY_ORDERS  * g * u
    A_revenue = LY_PROMO_DAILY_REVENUE * g * u * (aov / (1 + observed_aov_compression))
    B_orders  = TY_BASE_DAILY_ORDERS   * ly * u
    B_revenue = TY_BASE_DAILY_REVENUE  * LY_REVENUE_UPLIFT * u * aov
    daily_rev = bw * A_revenue + (1 - bw) * B_revenue
    return daily_rev * PROMO_DAYS

print("=== SENSITIVITY ANALYSIS — IMPACT ON TOTAL REVENUE ===")
print(f"Base (mid) revenue: ${BASE:,.0f}")
print()
print(f"{'Assumption':<25} {'Base val':>10} {'-10%':>12} {'+10%':>12} {'Range':>12} {'Rank':>6}")
print("-" * 78)

sensitivities = []
for name, (param, base_val, pct) in assumptions.items():
    lo_rev = run_single(param, base_val * (1 - pct))
    hi_rev = run_single(param, base_val * (1 + pct))
    rng    = hi_rev - lo_rev
    sensitivities.append((name, base_val, lo_rev, hi_rev, rng))

sensitivities.sort(key=lambda x: -x[4])
for rank, (name, bv, lo, hi, rng) in enumerate(sensitivities, 1):
    print(f"{name:<25} {bv:>10.3f} ${lo:>10,.0f} ${hi:>10,.0f} ${rng:>10,.0f} {rank:>5}")

print()
print(f"Most sensitive assumption: {sensitivities[0][0]}")
print(f"  A ±10% change swings revenue by ${sensitivities[0][4]:,.0f}")
````

## 9. Explicit Scenario Drivers

What would have to be true for each scenario to materialise?


````python
drivers = {
    "Low (1.10x growth)": [
        "DTC growth rate decelerates as Vitamin Shoppe / Target retail expansion captures new customers",
        "Promo fatigue: customers have seen SPRING20 before and respond with smaller baskets",
        "Competitive sitewide promos from peers reduce Cymbiotika's incremental lift",
        "Lower email/SMS list engagement than last year",
    ],
    "Mid (1.18x growth)": [
        "DTC grows at the assignment-provided rate — consistent with residual DTC share of 1.50x total growth",
        "Promo repeats LY uplift performance (same codes, same duration, similar awareness)",
        "AOV compresses by ~15-16% — same as observed this year — as customers buy more units at a discount",
        "No major external disruptions (competitor launches, supply issues)",
    ],
    "High (1.25x growth)": [
        "DTC outperforms as brand awareness compounds and email/SMS list grows YoY",
        "Promo generates stronger-than-LY lift due to improved marketing execution",
        "Basket sizes grow (higher UPT) offsetting AOV compression better than expected",
        "Retail presence (Target, Vitamin Shoppe) drives halo effect — more people aware of brand DTC",
    ],
}

for scenario, bullet_list in drivers.items():
    print(f"=== {scenario} ===")
    for b in bullet_list:
        print(f"  • {b}")
    print()
````

## 10. Finance Summary — 3–5 Bullets

> This is what you would send to the Finance partner before the promo.


````python
mid = results["mid"]
lo  = results["low"]
hi  = results["high"]

order_lift_pct   = (mid["blended_daily_orders"]  / TY_BASE_DAILY_ORDERS  - 1) * 100
revenue_lift_pct = (mid["blended_daily_revenue"] / TY_BASE_DAILY_REVENUE - 1) * 100
aov_compression  = (mid["projected_AOV"] / TY_BASE_AOV - 1) * 100

print("=" * 70)
print("SPRING20 FORECAST — FINANCE BRIEF")
print("=" * 70)
print()
print(f"• Mid-case projects {mid['total_orders']:,.0f} orders and "
      f"${mid['total_revenue']:,.0f} in net sales over the 3-day event "
      f"({order_lift_pct:+.0f}% order lift and {revenue_lift_pct:+.0f}% revenue lift "
      f"vs the 8-week trailing baseline of ${TY_BASE_DAILY_REVENUE:,.0f}/day).")
print()
print(f"• Plausible range is {lo['total_orders']:,.0f}–{hi['total_orders']:,.0f} orders "
      f"and ${lo['total_revenue']:,.0f}–${hi['total_revenue']:,.0f} in net sales. "
      f"The spread reflects genuine uncertainty in YoY DTC growth (1.10–1.25x) "
      f"and uplift repeatability — not rounding.")
print()
print(f"• AOV is forecast to compress ~{abs(aov_compression):.0f}% vs baseline "
      f"(from ${TY_BASE_AOV:.2f} to ~${mid['projected_AOV']:.2f}) as the 20% discount "
      f"reduces revenue per order. This is partially offset by a modest basket-size "
      f"increase (UPT expected to tick up from {TY_BASE_UPT:.2f} to ~{TY_PROMO_UPT:.2f}). "
      f"Plan inventory accordingly.")
print()
print(f"• The forecast is anchored on two signals: last year's SPRING20 event "
      f"(derived at {LY_PROMO_ORDERS} orders / ${LY_PROMO_REVENUE:,.0f} revenue) "
      f"scaled by the 1.18x YoY growth rate, blended 60/40 with the TY baseline × "
      f"LY uplift factor ({LY_ORDER_UPLIFT:.1f}x orders). Neither anchor alone is "
      f"sufficient — the blend reduces exposure to anomalies in either signal.")
print()
print(f"• Top risk to the upside: stronger-than-expected halo from retail expansion "
      f"(Target, Vitamin Shoppe) driving incremental DTC awareness. "
      f"Top risk to the downside: DTC growth rate below 1.10x if retail channels "
      f"are cannibalising online demand faster than assumed. "
      f"Sensitivity analysis shows YoY growth rate is the single assumption that "
      f"moves revenue most — a ±10% change in the growth rate swings the forecast "
      f"by approximately ${{swing:,.0f}}.".format(
          swing=sensitivities[0][4] if 'sensitivities' in dir() else 2000))
print()
print("=" * 70)
````

## 11. Save Outputs

````python
out_rows = []
for s in ["low", "mid", "high"]:
    row = {"scenario": s}
    row.update(results[s])
    out_rows.append(row)

out_df = pd.DataFrame(out_rows).set_index("scenario")
out_df.to_csv("part5_forecast.csv")
print("Saved → part5_forecast.csv")
print()
print("=== FINAL FORECAST TABLE ===")
print(out_df[[
    "total_orders",
    "total_revenue",
    "projected_AOV",
    "blended_daily_orders",
    "blended_daily_revenue",
]].round(2).to_string())
````
