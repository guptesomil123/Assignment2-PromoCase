## 3. Column-by-Column Cleaning

Working on a copy so the raw DataFrame stays intact for comparison.


```python
df = df_raw.copy()
```

### 3a. Column Names

**DECISION:** Strip leading/trailing whitespace from all column headers.  
Reason: CSVs exported from Shopify or custom ETL pipelines sometimes include stray spaces. Normalising here prevents `KeyError` surprises downstream.


```python
before = df.columns.tolist()
df.columns = df.columns.str.strip()
after = df.columns.tolist()

changed = [(b, a) for b, a in zip(before, after) if b != a]
print(f"Columns renamed: {changed if changed else 'none (all clean)'}")
```

### 3b. `order_id`

**DECISION:** Strip whitespace; deduplicate keeping first occurrence.  
Reason: A downstream join on `order_id` will silently fail (or double-count) if there are duplicates. First occurrence wins — earlier rows are more likely to be the canonical record (later duplicates are often re-exports).


```python
df["order_id"] = df["order_id"].str.strip()

dupes = df.duplicated(subset="order_id")
print(f"Duplicate order_ids: {dupes.sum()}")

if dupes.sum() > 0:
    print("\nSample duplicates:")
    display(df[df.duplicated(subset='order_id', keep=False)].sort_values('order_id').head(10))
    df = df.drop_duplicates(subset="order_id", keep="first").reset_index(drop=True)
    print(f"\nAfter dedup: {len(df):,} rows")
```

### 3c. `order_date`

**DECISION:** Parse to `datetime64` using `format="mixed"` with `errors="coerce"`.  
Then drop rows where parsing failed (NaT).  
Reason: Any order without a valid date cannot be assigned to the baseline window or the promo window — it is unusable for every metric in Parts 1–5. Keeping it would silently bias averages.  

> Note: `infer_datetime_format=True` was deprecated in pandas 2.0. `format="mixed"` handles heterogeneous formats (ISO, US, DD/MM/YYYY) gracefully.


```python
df["order_date"] = pd.to_datetime(df["order_date"], format="mixed", errors="coerce")

nat_count = df["order_date"].isna().sum()
print(f"Unparseable dates (NaT): {nat_count}")

if nat_count > 0:
    print("\nRows with bad dates:")
    display(df[df["order_date"].isna()])
    df = df.dropna(subset=["order_date"]).reset_index(drop=True)
    print(f"Dropped {nat_count} rows. Remaining: {len(df):,}")
else:
    print("All dates parsed successfully.")

print(f"\nDate range: {df['order_date'].min().date()} → {df['order_date'].max().date()}")
```

### 3d. `customer_type`

**DECISION:** Strip whitespace, normalise to lowercase.  
Map known casing variants to canonical values (`new`, `returning`, `subscriber`).  
Flag — do NOT silently drop — any unexpected values.  
Reason: Dropping unknown types would change promo mix metrics. Finance needs to decide whether to reclassify or exclude; the notebook surfaces them rather than hiding them.


```python
df["customer_type"] = df["customer_type"].str.strip().str.lower()

VALID_TYPES = {"new", "returning", "subscriber"}
unexpected = df.loc[~df["customer_type"].isin(VALID_TYPES), "customer_type"].unique()

if len(unexpected):
    print(f"⚠️  Unexpected customer_type values found: {unexpected}")
    print(df[~df["customer_type"].isin(VALID_TYPES)][["order_id", "order_date", "customer_type"]].head(10))
else:
    print("✓  All customer_type values are in expected set.")

print()
print(df["customer_type"].value_counts())
```

### 3e. `discount_code`

**DECISION:** Strip whitespace, uppercase.  
Coerce empty strings, `"nan"`, `"None"`, and `"NULL"` → `NaN` (pandas `NA`).  
Reason: An empty string `""` and a null both mean "no code used." Mixed representation  
causes false negatives in `discount_code == 'SPRING20'` filters. Uppercasing prevents  
`spring20` / `Spring20` misses.


```python
df["discount_code"] = (
    df["discount_code"]
    .astype(str)
    .str.strip()
    .str.upper()
    .replace({"NAN": np.nan, "": np.nan, "NONE": np.nan, "NULL": np.nan})
)

print("discount_code value counts (post-clean):")
print(df["discount_code"].value_counts(dropna=False).to_string())
```

### 3f. `net_sales`

**DECISION:** Strip any `$` or `,` characters, then coerce to `float`.  
Keep negative values (refunds / cancellations) — flag them but don't drop.  
Drop rows where `net_sales` is null after coercion.  
Reason:  
- Negatives affect AOV and revenue totals; Finance should see them, not have them silently removed.  
- Nulls cannot contribute to any metric and would propagate NaN into aggregates.


```python
df["net_sales"] = (
    df["net_sales"]
    .astype(str)
    .str.replace(r"[$,]", "", regex=True)
    .str.strip()
    .pipe(pd.to_numeric, errors="coerce")
)

null_sales  = df["net_sales"].isna().sum()
neg_sales   = (df["net_sales"] < 0).sum()
zero_sales  = (df["net_sales"] == 0).sum()

print(f"null net_sales  : {null_sales}")
print(f"negative net_sales (refunds): {neg_sales}")
print(f"zero net_sales  : {zero_sales}")

if null_sales > 0:
    print("\nRows with null net_sales:")
    display(df[df["net_sales"].isna()].head(5))
    df = df.dropna(subset=["net_sales"]).reset_index(drop=True)
    print(f"Dropped {null_sales} rows. Remaining: {len(df):,}")
```

### 3g. `units`

**DECISION:** Coerce to `int`. Drop nulls. Keep rows with units ≤ 0 but flag them.  
Reason: UPT (units per transaction) requires valid unit counts. Zero or negative unit orders may represent cancellations — consistent with keeping negative net_sales.


```python
df["units"] = pd.to_numeric(df["units"], errors="coerce")

null_units = df["units"].isna().sum()
if null_units > 0:
    df = df.dropna(subset=["units"]).reset_index(drop=True)
    print(f"Dropped {null_units} rows with null units.")

df["units"] = df["units"].astype(int)

non_pos = (df["units"] <= 0).sum()
print(f"Orders with units ≤ 0: {non_pos}")
```

### 3h. `aov` and `upt` (Derived Columns)

**DECISION:** Recompute both from `net_sales` and `units` rather than trusting the raw values.  
Then check for discrepancies > $0.01 between raw and recomputed values.  
Reason: At the **order level**, `aov = net_sales` and `upt = units` for a single order.  
True AOV / UPT are only meaningful when aggregated across many orders.  
If the raw values differ from recomputed, that signals an upstream calculation error.


```python
df["aov_raw"] = pd.to_numeric(df["aov"], errors="coerce")
df["upt_raw"] = pd.to_numeric(df["upt"], errors="coerce")

df["aov"] = df["net_sales"]
df["upt"] = df["units"]

aov_diff = (df["aov_raw"] - df["aov"]).abs()
diff_rows = (aov_diff > 0.01).sum()

if diff_rows > 0:
    print(f"⚠️  {diff_rows} rows where raw aov ≠ net_sales (diff > $0.01):")
    display(df[aov_diff > 0.01][["order_id", "order_date", "net_sales", "aov_raw"]].head(10))
else:
    print("✓  aov and net_sales are consistent across all rows.")

df = df.drop(columns=["aov_raw", "upt_raw"])
```

### 3i. `skus`

**DECISION:** Strip whitespace from the raw string. Leave pipe-delimited as-is.  
Reason: Parts 5–6 will explode this column. Cleaning it here (strip only) avoids  
`CYM-PRO-30 ` (trailing space) being treated as a different SKU from `CYM-PRO-30`.


```python
df["skus"] = df["skus"].astype(str).str.strip()

# Sanity check: any SKU strings that look malformed?
has_pipe = df["skus"].str.contains("|", regex=False)
print(f"Multi-SKU orders (pipe-delimited): {has_pipe.sum():,} of {len(df):,}")
print()
print("Sample multi-SKU row:")
print(df.loc[has_pipe, "skus"].iloc[0] if has_pipe.any() else "none")
```

## 5. Promo Definition — SPRING20 Filter

Before building the summary table, we need to pin down exactly which orders count as "promo."

Three candidate definitions:

| Definition | Pros | Cons |
|---|---|---|
| **A** Date window only (Apr 25–27) | Simple | Includes organic orders on those days with no code |
| **B** Code only (`SPRING20`) | Captures any late-coded orders | Misses true promo window edge cases; code could be used outside the event |
| **C** Date window **AND** code ✅ | Most defensible for lift analysis | Slightly narrower than reality, but conservative |

**We use Definition C.** It is the standard approach for promo lift analysis: we want orders that were specifically driven by the SPRING20 discount within the event window. Mixing in organic same-day orders would inflate the "baseline" on promo days, flattening the lift calculation in Part 2.


```python
mask_promo = (
    (df["order_date"] >= PROMO_START) &
    (df["order_date"] <= PROMO_END) &
    (df["discount_code"] == PROMO_CODE)
)

promo = df[mask_promo].copy()
non_promo = df[~mask_promo].copy()

print(f"Promo orders (SPRING20, Apr 25–27) : {len(promo):,}")
print(f"Non-promo orders                   : {len(non_promo):,}")
print(f"Total                              : {len(df):,}")
print()

# Cross-check: any SPRING20 codes outside the window?
code_only = df[df["discount_code"] == PROMO_CODE]
outside_window = code_only[
    (code_only["order_date"] < PROMO_START) |
    (code_only["order_date"] > PROMO_END)
]
print(f"SPRING20 codes outside Apr 25–27 window: {len(outside_window)}")
if len(outside_window):
    print(outside_window[["order_id","order_date","discount_code","net_sales"]].to_string())
```

## 6. Promo Summary Table

Key metrics for the SPRING20 event.


```python
# Customer-type breakdown
ctype = promo["customer_type"].value_counts()
ctype_pct = promo["customer_type"].value_counts(normalize=True).mul(100).round(1)

print("=== PROMO ORDERS BY DAY ===")
print(promo.groupby(promo["order_date"].dt.date).agg(
    orders=("order_id", "count"),
    net_sales=("net_sales", "sum"),
    aov=("net_sales", "mean"),
    upt=("units", "mean")
).round(2).to_string())
print()
```

```python
# Build summary table
summary = {
    "Total orders"                    : len(promo),
    "Total net sales ($)"             : round(promo["net_sales"].sum(), 2),
    "AOV ($)"                         : round(promo["net_sales"].mean(), 2),
    "Median order value ($)"          : round(promo["net_sales"].median(), 2),
    "UPT (avg units per transaction)" : round(promo["units"].mean(), 2),
    "% New customers"                 : ctype_pct.get("new", 0),
    "% Returning customers"           : ctype_pct.get("returning", 0),
    "% Subscribers"                   : ctype_pct.get("subscriber", 0),
    "# unique SKUs ordered"           : promo["skus"].str.split("|").explode().str.strip().nunique(),
    "Orders using SPRING20 code"      : (promo["discount_code"] == PROMO_CODE).sum(),
}

summary_df = pd.DataFrame(
    [(k, v) for k, v in summary.items()],
    columns=["Metric", "Value"]
)

summary_df.to_csv("part1_promo_summary.csv", index=False)
print("Saved → part1_promo_summary.csv")
print()
summary_df
```

```python
# Subscriber vs one-time split (separate from customer_type column)
sub_orders    = (promo["customer_type"] == "subscriber").sum()
nonSub_orders = len(promo) - sub_orders
sub_rev       = promo.loc[promo["customer_type"] == "subscriber", "net_sales"].sum()
nonSub_rev    = promo["net_sales"].sum() - sub_rev

print("=== SUBSCRIBER vs ONE-TIME BREAKDOWN ===")
print(f"{'Segment':<20} {'Orders':>8} {'%':>6} {'Revenue ($)':>14} {'%':>7}")
print("-" * 60)
total_orders = len(promo)
total_rev    = promo["net_sales"].sum()
for seg, orders, rev in [
    ("Subscriber",    sub_orders,    sub_rev),
    ("One-time",      nonSub_orders, nonSub_rev),
]:
    print(f"{seg:<20} {orders:>8,} {orders/total_orders*100:>5.1f}%"
          f" {rev:>13,.2f} {rev/total_rev*100:>6.1f}%")
print("-" * 60)
print(f"{'TOTAL':<20} {total_orders:>8,} {'100.0':>6}% {total_rev:>13,.2f} {'100.0':>6}%")
```
